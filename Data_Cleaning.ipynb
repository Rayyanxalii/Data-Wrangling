```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/employee_data.csv')
data.head(10)
```

### Dirty Data

- Incorret and missing values in Name , Age, Gender, Salary, Location, Performance score



### Messy Data

- Employee id can be used as index
- In gender and department, the words are not formatted correctly
- Wrong format in join date
- Wrong data type for join date

```python
data.info()
```

```python
data_copy = data.copy()
```

```python
data_copy = data_copy.drop('Join_Date', axis=1)
```

```python
# Correcting Employee Id
data_copy['Employee_ID'] = data_copy['Employee_ID'] - 3000
data_copy['Employee_ID'] = data_copy['Employee_ID'].astype('int32')
data_copy.set_index('Employee_ID', inplace=True)
```

```python
# Correcting name column
data_copy.loc[data_copy['Full_Name'].str.startswith('Person_'), 'Full_Name'] = 'Unknown'
```

```python
# Correcting Age Column
data_copy['Age'] = data_copy['Age'].replace(999,np.nan)
data_copy['Age']= data_copy['Age'].fillna(data_copy['Age'].mean())
data_copy['Age'] = data_copy['Age'].astype('int16')
```

```python
# Correcting Gender Column
data_copy['Gender'] = data_copy['Gender'].replace('Fmale','Female')
data_copy['Gender'] = data_copy['Gender'].fillna(data_copy['Gender'].mode()[0])
data_copy['Gender'] = data_copy['Gender'].astype('category')
```

```python
# Correcting Department
data_copy['Department'] = data_copy['Department'].str.capitalize()
data_copy['Department'] = data_copy['Department'].replace('Hr', 'Human Resources')
data_copy['Department'] = data_copy['Department'].replace('It', 'Information Technology')
```

```python
data_copy['Salary'] = data_copy['Salary'].replace(0.0 , data_copy['Salary'].mean())
data_copy['Salary'] = data_copy['Salary'].fillna(data_copy['Salary'].mean())
data_copy['Salary'] = data_copy['Salary'].astype('int32')
```

```python
data_copy['Location'] = data_copy['Location'].str.title()
data_copy['Location'] = data_copy['Location'].fillna(data_copy['Location'].mode()[0])
```

```python
data_copy['Performance_Score'] = data_copy['Performance_Score'].fillna(data_copy['Performance_Score'].mean())
data_copy['Performance_Score'] = data_copy['Performance_Score'].astype('int16')
```
